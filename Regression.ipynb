```python
df = df[['Adj. Open',  'Adj. High',  'Adj. Low',  'Adj. Close', 'Adj. Volume']]
```

```python
df['HL_PCT'] = (df['Adj. High'] - df['Adj. Low']) / df['Adj. Close'] * 100.0
df['PCT_change'] = (df['Adj. Close'] - df['Adj. Open']) / df['Adj. Open'] * 100.0
df = df[['Adj. Close', 'HL_PCT', 'PCT_change', 'Adj. Volume']]
print(df.head())
```

```python
df
```

```python
forecast_column = 'Adj. Close'
df.fillna(value = -99999, inplace=True)      #fill if not a number
import math
forecast_out= int(math.ceil(0.01*len(df)))  
df['label'] = df[forecast_column].shift(-forecast_out)
# label is price in the future, where the future is 10% of the entire length of the dataset out.
print(df.head())
```

## Training and Testing

```python
import numpy as np
from sklearn import preprocessing, cross_validation, svm
from sklearn.linear_model import LinearRegression
```

```python
df.dropna(inplace=True)  #dropping NaN info still left
#X - features y- label
X = np.array(df.drop(['label'],1))  #everything except label
y = np.array(['label'])

X = preprocessing.scale(X)  #scaling X
y = np.array(df['label'])
```

```python
print(len(X), len(y))
```

```python
X_train, X_test, y_train , y_test= cross_validation.train_test_split(X,y,test_size=0.2)
#80 % to train rest for test
classifier = LinearRegression()
classifier.fit(X_train,y_train)
```

```python
accuracy = classifier.score(X_test,y_test)
print(accuracy)
```

```python
print(forecast_out)
```

```python
classifier1 = svm.SVR()    #Support Vector Regression
classifier1.fit(X_train,y_train)
accuracy1 = classifier1.score(X_test,y_test)
print(accuracy1)
```

```python
classifier1 = svm.SVR(kernel='poly')
classifier1.fit(X_train,y_train)
accuracy1 = classifier1.score(X_test,y_test)
print(accuracy1)
```

```python
classifier = LinearRegression(n_jobs=-1)  #training as fast as possible... eg. 20 doing 20 jobs at a time
classifier.fit(X_train,y_train)
accuracy = classifier.score(X_test,y_test)
print(accuracy)
```

```python
#for different kernels in the algorithm
for k in ['linear','poly','rbf','sigmoid']:
    classifier = svm.SVR(kernel=k)
    classifier.fit(X_train, y_train)
    confidence = classifier.score(X_test, y_test)
    print(k,confidence)
```

## Predicting

```python
#print(len(X), len(y))
df.dropna(inplace=True)  #dropping NaN info still left
#X - features y- label
X = np.array(df.drop(['label'],1))  #everything except label
y = np.array(['label'])

X = preprocessing.scale(X)  #scaling X
y = np.array(df['label'])
print(len(X), len(y))


X_lately = X[-forecast_out:]
#X = X[:-forecast_out]
print(len(X), len(y))

X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2)
clf = LinearRegression(n_jobs=-1)
clf.fit(X_train, y_train)
confidence = clf.score(X_test, y_test)
print(confidence)
```

```python
forecast_set = clf.predict(X_lately)
```

```python
print(forecast_set, confidence, forecast_out)
```
